```python
# Dependencies and dataload
# Link to associated webkit
# http://scikit-learn.org/stable/modules/feature_selection.html#l1-feature-selection

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
import matplotlib.pyplot as plt
import seaborn as sns

my_data = pd.read_csv('/Users/stevensun/Prometheus/0_data/prometheus.csv')
my_data.shape
```

```python
#drop incomplete rows

my_data = my_data.dropna()
my_data.shape
```

```python
# vectorize data

X = my_data.ix[:,0:125]
y = my_data.ix[:,126]
X_headers = X.columns
```

```python
##create linear regression model
lm = linear_model.LinearRegression()
lm.fit(X, y)
```

```python
# Drow row values = 0
# Order from least to most
features = pd.DataFrame(lm.coef_,X.columns,columns=['Coefficients'])
features = features[(features.T != 0).any()]
features.sort_values(by='Coefficients', ascending=1)
```

```python
# The mean squared error
print("Mean squared error: %.2f"
      % np.mean((lm.predict(X) - y) ** 2))

# Explained variance score: 1 is perfect prediction
print('Variance score: %.2f' % lm.score(X, y))
```
